```python

# Write a recursive function to generate the nth number in the fibonacci sequence (starting from 0)
# example, fib(1) should return 1
# fib(3) should return 3
# fib (7) should return 21 ...
# Function to find the nth Fibonacci number
def fib(n):
    if n == 0:
        return 1
    elif n == 1:
        return 1
    elif n == 2:
        return 2
    else:
        return fib(n-1) + fib(n-2)
    
    
n=7
print (fib(n))
```

```python

# Write a non-recursive function to generate the nth number in the fibonacci sequence (starting from 0)
# example, nr_fib(1) should return 1
# nr_fib(3) should return 3
# nr_fib (7) should return 21 ...

def nr_fib(n):
    arr_n = [0, 1]  # initialize with first two values
    
    for i in range(2, n+1):
        arr_n.append(arr_n[-1] + arr_n[-2])
    
    return arr_n[-1]+arr_n[-2]

print(nr_fib(3))
```

```python
#1- The implementation of the two functions is very different. 
# The recursive function uses a recursive approach to generate the nth Fibonacci 
#number, while the non-recursive function uses a loop to iteratively generate the 
#nth Fibonacci number.
#2- The recursive function is less efficient than the non-recursive function, 
#especially for large values of n. This is because the recursive function 
#generates the same sub-problems multiple times, leading to redundant computation. 
#In contrast, the non-recursive function only generates each sub-problem once.
#3- The recursive function uses more memory than the non-recursive function 
#because it stores each intermediate result on the call stack. This can be a problem 
#for large values of n, as it can cause the call stack to overflow. The non-recursive 
#function, on the other hand, only stores the previous two values in the Fibonacci 
#sequence at any given time, so it uses less memory.
```

```python
#Recursion to calculate factorial

def fact(n):
    if n == 0:
        return 1
    else:
        return n * fact(n-1)
    
    
fact(4)
```

```python
#Non recursive function to calculate factorial

def nr_fact(n):
    arr_n = [1 for i in range(n+1)]
    
    for i in range(2, n+1):
        arr_n[i] = i * arr_n[i-1]
        
    return arr_n[n]

nr_fact(7)
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def build_tree(nodes, i):
    if i >= len(nodes) or nodes[i] is None:
        return None
    root = TreeNode(nodes[i])
    root.left = build_tree(nodes, 2*i + 1)
    root.right = build_tree(nodes, 2*i + 2)
    return root

nodes = [5, 12, 9, 7, None, 18, 33]
root = build_tree(nodes, 0)
```
